```python
from sklearn import tree
```

```python
features=[[150,0],
[170,0],
[130,1],
[120,1]]
```

```python
labels=["apple","apple","orange","orange"]
```

```python
clf=tree.DecisionTreeClassifier()
```

```python
clf=clf.fit(features,labels)
```

```python
prediction=clf.predict([[150,1]])
```

```python
print(f"The fruit is {prediction[0]}")
```
